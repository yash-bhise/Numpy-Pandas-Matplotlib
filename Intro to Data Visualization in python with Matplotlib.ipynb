```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Basic Graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#resize your Graph
plt.figure(figsize=(6,4),dpi=100) #(dpi specifies pixels per inch,when saving probably should use 300) 

#Keyword Argument Notation
#plt.plot(x,y, label='2x',color='red',linestyle='--', linewidth=2, marker='.',#markersize=10,markeredgecolor='red')
    
# Shorthand notation
plt.plot(x,y, 'b^--' ,label='2x')

#Line no.2
# Select interval we want to plot points at
x2 = np.arange(0,4,0.5)

# Plot remainder of graph as line
plt.plot(x2[:6],x2[:6]**2,'r',label='X^2')
plt.plot(x2[5:],x2[5:]**2,'r--')

# Add a title,xlabel,ylabel
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('First Graph', 
          fontdict={'fontname': 'Comics Sans MS',
                    'fontsize':20})

# x,y axis tick marks
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi=300+ for better resolution)
plt.savefig('mygraph.png',dpi=300)

# Show plot
plt.show()
```

### Bar chart

```python
l=['A','B','C']
v=[1,6,3]

bars = plt.bar(l,v)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.figure(figsize=(10,4))

plt.show()
```
